```python
response = {'seq': '고유번호',
            'statedt': '기준일',
            'statetime': '기준시간',
            'decidecnt': '확진자',
            'clearcnt': '격리해제',
            'examcnt': '검사진행',
            'deathcnt': '사망자',
            'carecnt': '치료중',
            'resutlnegcnt': '음성',
            'accexamcnt': '누적검사',
            'accexamcompcnt': '누적검사완료',
            'accdefrate': '누적확진률',
            'createdt': '등록일시',
            'updatedt': '수정'}
```

```python
for x in items:
    print(x)
    print()
```

```python
lst = []
for y in items:
    l = {}
    for x in y:
        l[x.name]=x.text
    lst.append(l)
lst
```

```python
# list를 dataframe으로 만들기
df = pd.DataFrame(lst)
df.columns = df.columns.map(response)
df
```

```python
df['확진자']=df['확진자'].astype(int)
```

```python
decidecnt=[]
for index, val in enumerate(df['확진자']):
    if(index < len(df['확진자'])-1):
        decidecnt.append(df['확진자'][index]-df['확진자'][index+1])
```

```python
decidecnt = pd.Series(decidecnt)
```

```python
df['확진자'] = decidecnt
```

```python
type(df['확진자'][0])
df['확진자'][0]-df['확진자'][1]
```

```python
df['등록일시']=df['등록일시'].str.split(" ").str[0]
```

```python
df['등록일시'] = df['등록일시'].apply(pd.to_datetime)
```

```python
df_refined = df[['등록일시','확진자']].sort_values('등록일시',ascending=True).reset_index()
```
